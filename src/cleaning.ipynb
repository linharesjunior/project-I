```python
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
import seaborn as sns
```

```python
shark = pd.read_csv("/Users/juniorlinhares/Documents/projects/project-I/data/attacks.csv", encoding="Latin-1")
shark.sample()
```

```python
shark.columns
```

```python
missing_values = shark.isnull().sum()
missing_values
```

```python
# removing rows with no information in any columns

shark2 = shark.dropna(how='all')
shark2.reset_index(drop=True, inplace=True)
shark2
```

```python
# Columns that doesn’t contain useful information

shark3 = shark2.drop(['Case Number', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23'], axis=1)
shark3.sample()
```

```python
# Standardizing column names

shark3.columns = [i.lower().replace(" ", "_") for i in shark3.columns]
shark3.sample()
```

```python
# Converting date columns

shark3['date'] = pd.to_datetime(shark3['date'], errors='coerce')
shark3.sample()
```

```python
# Using regex101 to create

shark3['date'] = shark3['date'].astype(str)

pattern = r'\b\w+\s+(\d{4}-\d{2}-\d{2})'

shark3['date'] = shark3['date'].str.replace(pattern, '', regex=True)
shark3['date'] = pd.to_datetime(shark3['date'], errors='coerce')
shark3.sample()
```

```python
# Working on 'age' column, excluding letters and wrong text in the rows, using regex101!

shark3['age'] = shark3['age'].astype(str)
shark3['age'] = shark3['age'].str.extract('(\d+)')
shark3['age'] = pd.to_numeric(shark3['age'], errors='coerce')

shark3.sample()
```

```python
# Convert the 'year' column to integers

shark3['year'].fillna(-1, inplace=True)

shark3['year'] = shark3['year'].astype(int)
shark3.sample()
```

```python
# Convert the 'year' column to integers

shark3['year'] = pd.to_numeric(shark3['year'], errors='coerce')
shark3.sample()
```

```python
shark3.columns
```

```python
# removing duplicated rows 

shark3.drop_duplicates(inplace = True)
shark3.sample()
```

```python
# Using regex to replace the matched pattern with 'Boating'

pattern = r'.*Boat.*'
shark3['type'] = shark3['type'].str.replace(pattern, 'Boating', regex=True)
```

```python
# Export the data

shark3.to_csv("../data/cleansharkattack.csv")
```
