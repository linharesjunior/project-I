```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
import seaborn as sns
```

```python
shark3 = pd.read_csv(("../data/cleansharkattack.csv"))
shark3
```

```python
""" 
Geographical Analysis:
What are the top 10 countries with the highest number of recorded shark attacks?
"""
country_counts = shark3['country'].value_counts()
top_10_countries = country_counts.head(10)

print(top_10_countries)
```

```python
activity_counts = shark3['activity'].value_counts()
top_10_activities = activity_counts.head(10)

print(top_10_activities)
```

```python
activity_categories = {
    'Surfing': 'Water Sports',
    'Swimming': 'Swimming',
    'Fishing': 'Fishing',
    'Spearfishing': 'Fishing',
    'Bathing': 'Swimming',
    'Wading': 'Swimming',
    'Diving': 'Water Sports',
    'Standing': 'Swimming',
    'Snorkeling': 'Water Sports',
    'Scuba diving': 'Water Sports'
}
```

```python
""" 
Activity Analysis
What are the top 10 activities with the highest number of recorded shark attacks?
"""

# Using the top 10 activities the people were doing during the attack, separated in 3 Category (Water Sports, Swimiing, and Fishing)

activity_counts = shark3['activity'].value_counts()
top_10_activities = activity_counts.head(10)


category_counts = {
    'Water Sports': 0,
    'Swimming': 0,
    'Fishing': 0
}

for activity, count in activity_counts.items():
    category = activity_categories.get(activity, 'Other')
    if category in category_counts:
        category_counts[category] += count

categories = list(category_counts.keys())
counts = list(category_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color=['#2F8AC3', '#234990', '#0B72D7'])
plt.title('Shark Attacks by Activity Category')
plt.xlabel('Activity Category')
plt.ylabel('Number of Attacks')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
"""
Is there a correlation between the fatality rate and factors like the location of the attack,
the type of attack involved?
"""

shark3['activity_category'] = shark3['activity'].map(activity_categories)

filtered_data = shark3[shark3['fatal_(y/n)'].isin(['Y', 'N'])]
data = filtered_data[['fatal_(y/n)', 'activity_category']]

contingency_table = pd.crosstab(data['fatal_(y/n)'], data['activity_category'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="coolwarm")
plt.title("Correlation between Fatality and Activity Category")
plt.xlabel("Activity Category")
plt.ylabel("Fatality (Y/N)")

plt.show()
```

```python
""" 
Temporal Analysis:
How has the frequency of shark attacks changed over the years? (1900 to 2018)
"""

shark3_filtered = shark3[shark3['year'] >= 1900]

# Group the filtered data by 'year' and count the number of shark attacks per year
yearly_counts = shark3_filtered['year'].value_counts().sort_index()

# Create a line plot to visualize the change over the years
plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='line', marker='o')
plt.title('Shark Attacks Over the Years (Since 1900)')
plt.xlabel('Year')
plt.ylabel('Number of Shark Attacks')
plt.grid(True)
plt.show()
```

```python
# Using regex to replace the matched pattern with 'Boating'

pattern = r'.*Boat.*'
shark3['type'] = shark3['type'].str.replace(pattern, 'Boating', regex=True)
```

```python
"""
Attack Types:
What are the most common types of shark attacks (e.g., provoked, unprovoked, boating-related)?
"""

filtered_shark3 = shark3[~shark3['type'].isin(['Invalid', 'Questionable'])]

type_counts = filtered_shark3['type'].value_counts()

plt.figure(figsize=(10, 6))
type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Shark Attacks per Type')
plt.xlabel('Shark Attack Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
"""
Analyse the season,  time of the year in months
checking if there;s a pattern, season or time of the year it happens the attacks happen

"""

shark3['date'] = pd.to_datetime(shark3['date'], errors='coerce')
shark3['month'] = shark3['date'].dt.month

monthly_counts = shark3['month'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
monthly_counts.plot(kind='line', marker='o')
plt.title('Distribution of Shark Attacks by Month')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
"""
Fatalities:
What percentage of shark attacks result in fatalities? done


Also, proportion of fatal attack per country, also by activity
"""

# What percentage of shark attacks result in fatalities?

fatality_percentage = (shark3['fatal_(y/n)'].value_counts(normalize=True) * 100).get('Y', 0)

labels = ['Fatal', 'Non-Fatal']
sizes = [fatality_percentage, 100 - fatality_percentage]
explode = (0.1, 0) 

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=['#FF9999', '#99FF99'], autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title('Shark Attack Fatality Percentage')

plt.show()
```
